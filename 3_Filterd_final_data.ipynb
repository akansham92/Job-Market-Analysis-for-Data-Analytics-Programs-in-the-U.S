```python
# Specify the column names for which to extract unique values
columns_of_interest = [
    'Programming_Languages',
    'Visualization_Tools',
    'Databases_and_Big_Data_Tools',
    'ML_and_Data_Science_Libraries',
    'Cloud_Services'
]

# Define a function to extract unique values from each column, ignoring case and cleaning data
def extract_unique_values(column_data):
    unique_items = set()
    for entry in column_data.dropna():  # Ignore NaN values
        items = [item.strip() for item in entry.split(',')]  
        unique_items.update(items)  # Add to the set to ensure uniqueness
    return unique_items

# Extract unique values for each column of interest
unique_values = {col: extract_unique_values(data[col]) for col in columns_of_interest}

# Output the results
for col, values in unique_values.items():
    print(f"Unique values in '{col}':")
    for value in sorted(values):
        print(f"  - {value}")
    print("\n")
```

```python
import re

# Define a function to filter valid items based on a regex pattern
def filter_valid_items(text, pattern):
    if pd.isna(text):
        return ""
    matches = re.findall(pattern, text)  
    return ", ".join(set(matches))  

# Iterate over the specified columns and apply the corresponding regex filtering
for column, pattern in patterns.items():
    if column in data.columns:
        data[column] = data[column].apply(lambda x: filter_valid_items(x, pattern))

# View the processed data
#data[patterns.keys()].head(100)
```

```python
import re

# Define mapping for various terms to standardized education levels
education_mapping = {
    'bachelors': 'bachelor',
    'undergraduate': 'bachelor',
    'msc': 'master',
    'masters': 'master',
    'doctorate': 'phd',
    'ma': 'master',
    'mba': 'master',  # Assuming MBA is equivalent to a master's level
    'doctor': 'phd',
    'ba': 'bachelor'
}

# Function to standardize and select only the first occurrence in order
def select_first_education(cell):
    if not isinstance(cell, str):
        return None 
    
    # Normalize all words to the standard form
    cell = cell.lower()
    for key, value in education_mapping.items():
        cell = re.sub(rf'\b{key}\b', value, cell)
    
    # Define the priority order
    priority_order = ['bachelor', 'master', 'phd']
    
    # Find the first occurrence in the priority order
    for level in priority_order:
        if level in cell:
            return level
    return None  # Return None if none of the levels are found

# Apply the function to select only the first occurrence based on the order
data['Education'] = data['Education'].apply(select_first_education)

# Extract unique values after processing
unique_education_values = data['Education'].dropna().unique()
unique_education_values
```

```python
import pandas as pd
import re

# Mapping of English number words to their corresponding numeric values
number_map = {
    "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
    "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10,
    "eleven": 11, "twelve": 12, "thirteen": 13, "fourteen": 14, 
    "fifteen": 15, "sixteen": 16, "seventeen": 17, "eighteen": 18, 
    "nineteen": 19, "twenty": 20, "thirty": 30, "forty": 40, 
    "fifty": 50, "sixty": 60, "seventy": 70, "eighty": 80, 
    "ninety": 90
}

# Define a function to extract and convert text into numeric values
def extract_numeric_experience(entry):
    if pd.isna(entry):  # Ignore NaN values
        return None
    # Define a regular expression to match numbers and English number words
    pattern = r"(\d+|one|two|three|four|five|six|seven|eight|nine|ten|eleven|twelve|thirteen|fourteen|fifteen|sixteen|seventeen|eighteen|nineteen|twenty|thirty|forty|fifty|sixty|seventy|eighty|ninety)"
    matches = re.findall(pattern, entry.lower())  # Extract all matches (case insensitive)
    
    if matches:
        # Convert the first matched number or word to its numeric equivalent
        first_match = matches[0]
        return int(first_match) if first_match.isdigit() else number_map.get(first_match, None)
    return None  # Return None if no matches are found

# Apply the function: Modify the 'Experience' column to contain only numeric values
data['Experience'] = data['Experience'].apply(extract_numeric_experience)

# Print the unique numeric values in the 'Experience' column
experience_unique_numbers = set(data['Experience'].dropna().unique())
print("\nUnique numeric values in 'Experience':")
for value in sorted(experience_unique_numbers, key=int):
    print(f"- {value}")
```

```python
import re

# Define mapping to standardize work modes
work_mode_mapping = {
    'onsite': 'on-site',
    'on-site': 'on-site',
    'in office': 'on-site',
    'in-office': 'on-site',
    'on site': 'on-site',
    
    'hybrid': 'hybrid',
    'hybrid model': 'hybrid',
    'flexible work': 'hybrid',
    
    'remote': 'remote',
    'remotely': 'remote',
    'virtual': 'remote',
    'work from home': 'remote',
    'wfh': 'remote'
}

# Function to standardize and filter work modes
def standardize_work_mode(cell):
    if not isinstance(cell, str):
        return None  
    
    # Normalize all words in the cell to standard form
    cell = cell.lower()
    for key, value in work_mode_mapping.items():
        cell = re.sub(rf'\b{key}\b', value, cell)
    
    # Extract only the relevant categories
    modes = {'on-site', 'hybrid', 'remote'}
    result = sorted(set(word for word in cell.split(', ') if word in modes))
    
    # If there is more than one mode, simplify to 'hybrid'
    if len(result) > 1:
        return 'hybrid'
    elif result:
        return result[0] 
    else:
        return None  

# Apply the standardization function
data['Work_Mode'] = data['Work_Mode'].apply(standardize_work_mode)

# Extract unique values after processing
Work_Mode_unique_values = data['Work_Mode'].dropna().unique()
Work_Mode_unique_values
```

```python
import pandas as pd
# Define classification logic
def classify_title(title):
    title = title.lower()
    if 'machine learning' in title or 'ai' in title or 'deep learning' in title or 'ml' in title:
        return "Machine Learning Engineer"
    elif 'data engineer' in title or 'etl' in title or 'big data' in title or 'snowflake' in title or 'data platform' in title:
        return "Data Engineer"
    elif 'architect' in title:
        return "Data Architect"
    elif 'analyst' in title:
        return "Data Analyst"
    elif 'science' in title or 'scientist' in title:
        return "Data Scientist"
    elif 'visualization' in title:
        return "Data Visualization"
    else:
        return "Other"

# Apply classification
data['job_position'] = data['Title'].apply(classify_title)
```

```python
# Define keywords for job levels and their corresponding categories
levels = {
    "Entry-level": ["Junior", "Entry", "Intern", "Assistant", "Associate"],
    "Senior-level": ["Senior", "Sr.", "Lead", "Principal", "Staff", "Specialist"],
    "Executive-level": ["Director", "VP", "Vice President", "Chief", "Head"]
}

# Function to extract the job level
def extract_level(title):
    for level, keywords in levels.items():
        if any(re.search(rf'\b{keyword}\b', title, re.IGNORECASE) for keyword in keywords):
            return level
    return "Unspecified"

# Apply the function to extract job levels
data["level"] = data["Title"].apply(extract_level)
```

```python
import pandas as pd
import re

# Cleaning function
def clean_location(location):
    if pd.isna(location):
        return None
    location = re.sub(r"\(.*?\)", "", str(location)).strip()  # Remove parentheses and their content
    if location.lower() in ["anywhere", "..."]:  # Clean special values
        return None
    return location

# Apply the cleaning function
data['Cleaned_Location'] = data['Location'].apply(clean_location)

# Safe splitting function
def safe_split(location):
    if pd.isna(location):
        return [None, None]
    parts = location.split(', ')
    if len(parts) == 2:
        return parts
    else:
        return [location, None]

# Apply safe splitting
data[['City', 'State']] = data['Cleaned_Location'].apply(safe_split).apply(pd.Series)

# Fill missing values
data['State'].fillna("Unknown", inplace=True)

# View the results
data.head()
```

```python
# Define a mapping table, including only inconsistent values that need to be processed
state_mapping = {
    'California': 'CA',
    'New York': 'NY',
    'Virginia': 'VA',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Oregon': 'OR',
    'North Carolina': 'NC',
    'Connecticut': 'CT',  # New mapping
    'Unknown': 'Unknown'
}

# Standardization function
def standardize_location(location):
    # If it's a two-character state abbreviation, return it directly
    if len(location) == 2:
        return location
    # Otherwise, check the mapping table
    return state_mapping.get(location, 'Unknown')

# Apply the function to the State column
data['State'] = data['State'].apply(standardize_location)
```
